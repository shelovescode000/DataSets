```python
#import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#load data
cab_data = pd.read_csv('Cab_Data.csv',',')
city_data = pd.read_csv('City.csv',',') #san francisco ca
customer_data = pd.read_csv('Customer_ID.csv',',')
trans_data = pd.read_csv('Transaction_ID.csv',',')
```

```python
#merge cab and city the merge trans and customer
cab_city = pd.merge(cab_data,city_data,on='City',how='left')
trans_customer = pd.merge(trans_data,customer_data,on='Customer ID',how='left')
```

```python
#merge all the data sets - some transaction in the transaction.csv dont have demographic data.
cab_fulldata = pd.merge(cab_city,trans_customer,on='Transaction ID',how='left')
cab_fulldata['Date of Travel'] =pd.to_datetime(cab_fulldata['Date of Travel'])
cab_fulldata.columns = ['Transaction ID', 'Date of Travel', 'Company', 'City', 'KMTravelled',
       'Price Charged', 'Cost of Trip', 'Population', 'Users', 'Customer ID',
       'Payment_Mode', 'Gender', 'Age', 'Income']
```

```python
cab_fulldata.info()
```

```python
cab_fulldata.head()
```

```python
#calculate profit from each trip
cab_fulldata['profit'] = cab_fulldata['Price Charged'] - cab_fulldata['Cost of Trip'] 
```

```python
#load the holiday dates
holidaydates = pd.read_excel('2010-2020 Holiday Dates.xlsx')
```

```python
cab_holiday_fulldata = pd.merge(cab_fulldata,holidaydates,on='Date of Travel',how='left')
```

```python
#replace the non holiday days
cab_holiday_fulldata['HOLIDAY_NAME']=cab_holiday_fulldata['HOLIDAY_NAME'].fillna('Non-holiday')
```

```python
#manipulating income variable
details = cab_holiday_fulldata['Income'].describe()
lowerQ = details['25%']
mean = details["mean"]
upperQ = details['75%']

# Generate result using pandas 
result = [] 
for value in cab_holiday_fulldata["Income"]: 
    if value < lowerQ: 
        result.append("low") 
    elif value<mean and value > lowerQ:
        result.append("middle") 
    else: 
        result.append("high") 
cab_holiday_fulldata["incomerange"] =result
```

```python
#separate holiday column to normal days and holidays
cab_holiday_fulldata["Holidays"] = cab_holiday_fulldata["HOLIDAY_NAME"].map(lambda x: "Normal" if "Non-holiday" in x  else "Holiday")
```

```python
#manipulating age groups
bins = [15, 25,35, 45,55,75]
labels = ['16-24', '25-34', '35-44', '45-54', '55+']
cab_holiday_fulldata['agerange'] = pd.cut(cab_holiday_fulldata.Age, bins, labels = labels,include_lowest = True)
```

```python
#seperate the date to days, months and years
cab_holiday_fulldata['weekday'] = pd.DatetimeIndex(cab_holiday_fulldata['Date of Travel']).weekday
cab_holiday_fulldata['month'] = pd.DatetimeIndex(cab_holiday_fulldata['Date of Travel']).month
cab_holiday_fulldata['year'] = pd.DatetimeIndex(cab_holiday_fulldata['Date of Travel']).year
```

```python
#manipulate the distance and place it as categories of short,long and normal
bins = [0, 10,20, 50]
labels = ['short', 'normal', 'long']
cab_holiday_fulldata['KMTravelled'] = pd.cut(cab_holiday_fulldata.KMTravelled, bins, labels = labels,include_lowest = True)
```

```python
#save the new data
cab_holiday_fulldata.to_csv(r'finaldata.csv', index = False)
```

```python
#Average Profit of the individual companies every
yearly_profits = cab_holiday_fulldata.groupby(['year', 'Company'])['profit'].mean().unstack().plot.bar(figsize=(12,8),color = ["pink", "yellow"])
plt.title("Yearly average profits for each company",fontsize= 16)

plt.xlabel('Year')
plt.ylabel('Average Profit (USD)')
for p in yearly_profits.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    yearly_profits.annotate(f'{round(height,2)}', (x + width/2, y + height*1.02), ha='center')
    
yearly_profits
```

```python
#Customer retention over the years
g = cab_holiday_fulldata.groupby(cab_holiday_fulldata.Company)
Yellow_cab = g.get_group('Yellow Cab')
Pink_cab = g.get_group('Pink Cab')

yellow_year = Yellow_cab.groupby(['year','Customer ID'])['Transaction ID'].size().unstack()
pink_year = Pink_cab.groupby(['year','Customer ID'])['Transaction ID'].size().unstack()

yr = (yellow_year.sum(axis=0,skipna=True))/3
pr = (pink_year.sum(axis=0,skipna=True))/3

bins = [0,10,40]
labels = ['bad', 'good']
Yellow_cab['retention'] = pd.cut(yr, bins, labels = labels,include_lowest = True)
Pink_cab['retention'] = pd.cut(pr, bins, labels = labels,include_lowest = True)
```

```python
Company_retentionY = Yellow_cab.groupby(['Company', 'retention']).size().unstack().plot.bar(figsize=(6,4))
plt.title("Yellow Cab:Retention vs Company",fontsize= 16)
plt.xlabel('Company')
plt.ylabel('Retention')
plt.xticks(rotation=0)
Company_retentionY
```

```python
Company_retentionP = Pink_cab.groupby(['Company', 'retention']).size().unstack().plot.bar(figsize=(6,4))
plt.title("Pink Cab: Retention vs company",fontsize= 16)
plt.ylabel('Retention')
plt.xticks(rotation=0)
Company_retentionP
```
